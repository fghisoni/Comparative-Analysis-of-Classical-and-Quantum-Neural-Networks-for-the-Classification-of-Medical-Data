### Amplitude encoding

```python
lr = 0.001
batch_size = 50
stoc_step = 5
epochs = 1000
layers = 7

amp_mse_test_accs_mean = []
amp_mse_test_accs_std = []

amp_msedr_test_accs_mean = []
amp_msedr_test_accs_std = []

amp_ce_test_accs_mean = []
amp_ce_test_accs_std = []

amp_cedr_test_accs_mean = []
amp_cedr_test_accs_std = []

amp_layer_mean = []
amp_layer_std = []
```

```python
for l in range(1,layers+1):
    int_list_mse = np.zeros((1000,10))
    int_list_msedr = np.zeros((1000,10))
    int_list_ce = np.zeros((1000,10))
    int_list_cedr = np.zeros((1000,10))

    amp_res = []
    for seed in range(1,11):
        fname_mse = f'Amplitude_Results_FINAL/MSE/test_acc_{l}layers_lerningrate{lr}_batchsize{batch_size}_stocstep{stoc_step}_{epochs}epochs_{seed}seed_MSE.txt'
        fname_msedr = f'Amplitude_Results_FINAL/MSE_DataReuploading/test_acc_{l}layers_lerningrate{lr}_batchsize{batch_size}_stocstep{stoc_step}_{epochs}epochs_{seed}seed_MSE_DataReuploading.txt'
        fname_ce = f'Amplitude_Results_FINAL/Cross_Entropy/test_acc_{l}layers_lerningrate{lr}_batchsize{batch_size}_stocstep{stoc_step}_{epochs}epochs_{seed}seed_CrossEntropy.txt'
        fname_cedr = f'Amplitude_Results_FINAL/Cross_Entropy_DataReuploading/test_acc_{l}layers_lerningrate{lr}_batchsize{batch_size}_stocstep{stoc_step}_{epochs}epochs_{seed}seed_Cross_Entropy_DataReuploading.txt'
    
        with open(fname_mse, 'r') as f:
            mse_vals = [float(i) for i in f.readlines()]
            int_list_mse[:,seed-1] = np.copy(mse_vals)
            # amp_res.append(mse_vals[-1])
        with open(fname_msedr, 'r') as f:
            msedr_vals = [float(i) for i in f.readlines()]
            int_list_msedr[:,seed-1] = np.copy(msedr_vals)
            amp_res.append(msedr_vals[-1])
        with open(fname_ce, 'r') as f:
            ce_vals = [float(i) for i in f.readlines()]
            int_list_ce[:,seed-1] = np.copy(ce_vals)
            # amp_res.append(ce_vals[-1])
        with open(fname_cedr, 'r') as f:
            cedr_vals = [float(i) for i in f.readlines()]
            int_list_cedr[:,seed-1] = np.copy(cedr_vals)
            # amp_res.append(cedr_vals[-1])
    
    amp_mse_test_accs_mean.append([np.mean(int_list_mse[i,:]) for i in range(1000)])
    amp_mse_test_accs_std.append([1.96*np.std(int_list_mse[i,:])/np.sqrt(10) for i in range(1000)])

    amp_msedr_test_accs_mean.append([np.mean(int_list_msedr[i,:]) for i in range(1000)])
    amp_msedr_test_accs_std.append([1.96*np.std(int_list_msedr[i,:])/np.sqrt(10) for i in range(1000)])

    amp_ce_test_accs_mean.append([np.mean(int_list_ce[i,:]) for i in range(1000)])
    amp_ce_test_accs_std.append([1.96*np.std(int_list_ce[i,:])/np.sqrt(10) for i in range(1000)])

    amp_cedr_test_accs_mean.append([np.mean(int_list_cedr[i,:]) for i in range(1000)])
    amp_cedr_test_accs_std.append([1.96*np.std(int_list_cedr[i,:])/np.sqrt(10) for i in range(1000)])

    amp_layer_mean.append(np.mean(np.array(amp_res)))
    amp_layer_std.append(np.std(np.array(amp_res)))
```
